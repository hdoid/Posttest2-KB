**NOMOR 4:** code diatas menunjukan record yang bertipe string atau objek dan menampilkan jumlah dari record atribut label. 

```python
import matplotlib.pyplot as plt
import seaborn as sns

hitunglabel = dataDio053['Future Career'].value_counts()

plt.figure(figsize=(10,6))
sns.barplot(x=hitunglabel.index, y=hitunglabel.values)

plt.title("Comp. Science Student Future Career", fontsize=16)
plt.xlabel("Future Career", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.xticks(rotation=45, ha='right')  # Memiringkan label x agar lebih mudah dibaca
plt.tight_layout()

plt.show()
```

**NOMOR 5:** untuk pertama hasil visualisasi diatas saya menggunakan diagram batang dimana atribut yang ditunjukan adalah future career.

```python
hitunglabel = dataDio053['Future Career'].value_counts()

plt.figure(figsize=(8,8))
plt.pie(hitunglabel.values, labels=hitunglabel.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title("Comp. Science Student Future Career", fontsize=16)

plt.show()
```

**NOMOR 5:** untuk kedua hasil visualisasi diatas saya menggunakan pie chart atau diagram lingkaran dan masih sama seperti sebelumnya atribut yang ditunjukan adalah future career.

```python
kolomtipeangka = dataDio053.select_dtypes(include=['number'])
sns.heatmap(kolomtipeangka.corr(), annot=True, annot_kws={'size':16})

plt.show()
```
